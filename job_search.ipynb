```python
import requests

def get_jobs_from_google(job_title, location="India"):
    api_key = "e480bb8e63f4ff99a46b1991d71131b7ae28b35bcd673c15499afb48482b3725"  
    params = {
        "engine": "google_jobs",
        "q": job_title,
        "location": location,
        "hl": "en",
        "api_key": api_key
    }

    # Send the request to SerpAPI
    response = requests.get("https://serpapi.com/search", params=params)
    
    # Check the response status code and print any error if exists
    if response.status_code != 200:
        print(f"Error: Unable to fetch data (status code: {response.status_code})")
        return []
    
    data = response.json()
    
    # Extract jobs results
    jobs = data.get("jobs_results", [])
    
    if not jobs:
        print("No jobs found.")
        return []

    # Cleaned up job data to print only relevant information
    job_data = []
    for idx, job in enumerate(jobs, 1):
        title = job.get('title', 'N/A')
        company = job.get('company_name', 'N/A')
        location = job.get('location', 'N/A')
        posted_at = job.get('detected_extensions', {}).get('posted_at', 'N/A')
        apply_link = job.get('link', 'N/A')

        # Only include relevant details
        job_data.append({
            "Job Title": title,
            "Company": company,
            "Location": location,
            "Posted": posted_at,
            "Apply Link": apply_link
        })

    return job_data

def display_jobs(job_data):
    if not job_data:
        print("No job data to display.")
        return
    
    print(f"\nTop {len(job_data)} Job Listings:")
    print("-" * 50)
    
    for idx, job in enumerate(job_data, 1):
        print(f"{idx}. {job['Job Title']} at {job['Company']}")
        print(f"   Location: {job['Location']}")
        print(f"   Posted: {job['Posted']}")
        print(f"   Apply Link: {job['Apply Link']}")
        print("-" * 50)

# Get job title from the user and fetch the job listings
job_title = input("Enter a job title: ")
job_listings = get_jobs_from_google(job_title)

# Display the cleaned job listings
display_jobs(job_listings)
```

```python
import requests

def get_jobs_from_google(job_title, location="India"):
    api_key = "e480bb8e63f4ff99a46b1991d71131b7ae28b35bcd673c15499afb48482b3725" 
    params = {
        "engine": "google_jobs",
        "q": job_title,
        "location": location,
        "hl": "en",
        "api_key": api_key
    }

    response = requests.get("https://serpapi.com/search", params=params)
    data = response.json()

    jobs = data.get("jobs_results", [])[:10]

    if not jobs:
        print("No jobs found.")
        return

    for idx, job in enumerate(jobs, 1):
        print(f"{idx}. {job.get('title')} at {job.get('company_name')}")
        print(f"   Location: {job.get('location')}")
        print(f"   Posted: {job.get('detected_extensions', {}).get('posted_at', 'N/A')}")
        
        apply_link = job.get('via', None)
        if 'link' in job:
            print(f"   Apply link: {job['link']}")
        else:
            print(f"   Apply via: {apply_link if apply_link else 'N/A'}")

      
        print(f"   Snippet: {job.get('description', '')[:120]}...")
        print("-" * 80)


job_title = input("Enter a job title: ")
get_jobs_from_google(job_title)
```
