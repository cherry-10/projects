```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import json

# Function to fetch job data from Google using SerpAPI
def get_jobs_from_google(job_title, location="India"):
    api_key = "e480bb8e63f4ff99a46b1991d71131b7ae28b35bcd673c15499afb48482b3725"  
    params = {
        "engine": "google_jobs",
        "q": job_title,
        "location": location,
        "hl": "en",
        "api_key": api_key
    }

    response = requests.get("https://serpapi.com/search", params=params)
    data = response.json()

    jobs = data.get("jobs_results", [])[:10]  # Limit to top 10 results

    if not jobs:
        print("No jobs found.")
        return []

    job_data = []
    for job in jobs:
        job_info = {
            "title": job.get('title', 'N/A'),
            "company": job.get('company_name', 'N/A'),
            "location": job.get('location', 'N/A'),
            "posted_at": job.get('detected_extensions', {}).get('posted_at', 'N/A'),
            "apply_links": [apply_option['link'] for apply_option in job.get('apply_options', [])],
            "skills": job.get('skills', []),  # List of skills if available
        }
        job_data.append(job_info)

    return job_data

# Get job title from the user and fetch the job listings
job_title = input("Enter a job title: ")
job_listings = get_jobs_from_google(job_title)

if job_listings:

    # Print the job listings (without description)
    print("\nJob Listings:")
    for idx, job in enumerate(job_listings, 1):
        print(f"{idx}. {job.get('title')} at {job.get('company_name')}")
        print(f"   Location: {job.get('location')}")
        print(f"   Posted: {job.get('posted_at', 'N/A')}")
        
        # Print all available apply links
        if job.get('apply_links'):
            print(f"   Apply links: {', '.join(job['apply_links'])}")
        else:
            print(f"   Apply link: N/A")
        
        print(f"   Skills: {', '.join(job.get('skills', []))}")
        print("-" * 80)
```
